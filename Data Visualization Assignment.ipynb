## Transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib
```

```python
#LOADING DATASET
df=sns.load_dataset('iris')
```

```python
df
```

```python
#FEATURE COMPONENT
x=df.iloc[:,:-1]
```

```python
x
```

```python
# Feature scaling before PCA
from sklearn.preprocessing import StandardScaler
st_sc=StandardScaler()
```

```python
X=st_sc.fit_transform(x)
```

```python
X
```

```python
# Reducing the dimensions to 3
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
X=pca.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder
Y=df['species']
le=LabelEncoder()
Y=le.fit_transform(Y)
colors=['red','green','blue']
```

```python
#Visualizing
fig=plt.figure(figsize=(16,10))
ax=plt.axes(projection='3d')
x=X[:,0]
y=X[:,1]
z=X[:,2]
ax.scatter(x,y,z,c=Y,cmap=matplotlib.colors.ListedColormap(colors))
ax.set_xlabel('1D')
ax.set_ylabel('2D')
ax.set_zlabel('3D')

plt.show()
```
